```python
import numpy as np
from sklearn import preprocessing
```

```python
def cost(x,y,m):
    ans = 0
    for i in range(len(x)):
        ans += ((y[i] - (m*x[i]).sum() )**2)
    return ans/len(x)
```

```python
def step_gd(x,y,m,al):
    m_s = np.zeros(len(x[0]))
    N = len(x)
    for i in range(N):
        for j in range(len(x[0])):
            m_s[j]+= (-2/N) * ( y[i] - (m*x[i]).sum() )*x[i][j]
        m = m - al*m_s
          
    
    return m
```

```python
def gd(x,y,al,num_it,m):
        
        for i in range(num_it):
            m = step_gd(x,y,m,al)
            print('cost',cost(x,y,m))
        return m
```

```python
def run1():
    data = np.loadtxt('train.csv',delimiter = ',')
    x = data[:,:-1]
    ones = np.ones((len(data),1))
    scaler = preprocessing.StandardScaler()
    scaler.fit(x)
    x = scaler.transform(x)
    
    x = np.append(x,ones,axis = 1) ## appended 1 for c
    y = data[:,-1]
    al = 0.005
    num_it = 1000
    m = np.zeros(len(x[0]))
    
    m = gd(x,y,al,num_it,m)

    xTest = np.loadtxt('test.csv',delimiter = ',')
    xTest = scaler.transform(xTest)
    xTest = np.append(xTest,np.ones((len(xTest),1)),axis = 1)
    
    y_ans = np.zeros(len(xTest))
    
    for i in range(len(xTest)):
        y_ans[i] = (m*xTest[i]).sum()
    
    np.savetxt('ans.csv',y_ans,delimiter = ',',fmt = '%.5f')
    
    
```

```python
run1()
```
